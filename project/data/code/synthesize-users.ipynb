```python
import pandas as pd
import random
```

```python
first_names = ["Aarav", "Vivaan", "Aditya", "Vihaan", "Arjun", "Sai", "Reyansh","Krishna", "Ishaan", "Rohan", "Ananya", "Diya", "Sanya", "Kavya","Meera", "Ira", "Riya", "Pooja", "Sneha", "Neha"]

last_names = ["Sharma", "Verma", "Gupta", "Singh", "Kumar", "Mehta","Agarwal", "Jain", "Patel", "Reddy", "Nair", "Iyer"]
```

```python
def random_name():
    return f"{random.choice(first_names)} {random.choice(last_names)}"
```

```python
def random_phone():
    return f"9{random.randint(100000000, 999999999)}"
```

```python
def random_password():
    base = ["Pass", "User", "Secure", "Login"]
    return random.choice(base) + str(random.randint(100, 999))
```

```python
def generate_users():
    users = []

    # --- 1. Students ---
    for i in range(1, 21):
        users.append({
            "user_id": f"STU{i:05}",
            "password": random_password(),
            "name": random_name(),
            "phone": random_phone(),
            "role": "student",
            "wing": None
        })

    # --- 2. Faculty ---
    for i in range(1, 9):
        users.append({
            "user_id": f"FAC{i:03}",
            "password": random_password(),
            "name": f"Dr. {random_name()}",
            "phone": random_phone(),
            "role": "faculty",
            "wing": None
        })

    # --- 3. Office/Admin Staff ---
    for i in range(1, 7):
        users.append({
            "user_id": f"OFF{i:03}",
            "password": random_password(),
            "name": random_name(),
            "phone": random_phone(),
            "role": "staff",
            "wing": None
        })

    # --- 4. Maintenance Staff ---
    wings = ["Plumbing", "Electrical", "IT", "Carpentry", "Cleaning"]
    
    mnt_id = 1
    for wing in wings:
        for _ in range(2):  # 2 staff per wing
            users.append({
                "user_id": f"MNT{mnt_id:03}",
                "password": "fixit" + str(random.randint(10, 99)),
                "name": random_name(),
                "phone": random_phone(),
                "role": "maintenance",
                "wing": wing
            })
            mnt_id += 1

    # --- 5. Super Admin ---
    users.append({
        "user_id": "ADMIN001",
        "password": "Root@123",
        "name": "System Administrator",
        "phone": random_phone(),
        "role": "superadmin",
        "wing": "All"
    })

    # --- Convert to DataFrame ---
    df = pd.DataFrame(users)
    df.to_csv('../users.csv', index=False)

    print(f"Generated {len(df)} realistic users")
    return df
```

```python
df_users = generate_users()
print(df_users.head())
```
